### Line Plots
Purpose: Showing trends in data -- usually time series data with many time points.

**matplotlib "Plot" Parameters-**

_color:_ Set the color of the line.

_linestyle:_ Set the line style, e.g., solid, dashed, or none. Options: https://matplotlib.org/examples/lines_bars_and_markers/line_styles_reference.html

_linewidth:_ Set the line thickness.

_marker:_ Set the marker style, e.g., circles, triangles, or none. Many different options: https://matplotlib.org/api/markers_api.html

_markersize:_ Set the marker size.

_label:_ Set the label for the line that will show up in the legend.

```python
x = np.linspace(0, 10)
y1 = np.sin(x)
y2 = np.sin(x - np.pi)

plt.figure()

# Same plot as before with tweaked parameters
plt.plot(x,
         y1,
         color='blue',
         linestyle='--',
         linewidth=2,
         marker='s',
         markersize=6,
         label='Plot 1')

plt.plot(x,
         y2,
         color='purple',
         linestyle='-',
         linewidth=2,
         marker='^',
         markersize=4,
         label='Plot 2')
;
```

### Bar Charts (Vertical and Horizontal)
### Vertical
Purpose: Comparing categories OR showing temporal trends in data with few (< 4) time points.

matplotlib function: bar(left, height)

left: The x coordinate(s) of the left sides of the bars.

height: The height(s) of the bars.

### Horizonal 

Purpose: Comparing categories.

matplotlib function: barh(bottom, width)

bottom: The y coordinate(s) of the bars.

width: The width(s) of the bars.

**matplotlib "Barh" and "Bar" Parameters-**

_color:_ Set the color of the bars.

_edgecolor:_ Set the color of the lines on the edges of the bars.

_height:_ Set the height of the bars.

_align:_ Set the alignment of the bars, e.g., center them on the y coordinate(s).

_label:_ Set the label for the bar that will show up in the legend.

```python
## Consider a case in looking at average transaction amounts by customer segment ##

categories = ['Customer Segment 1', 'Customer Segment 2', 'Customer Segment 3']
values = [30000, 4000, 10000]

plt.figure()

plt.barh(-np.arange(len(categories)),
         values,
         color='blue',
         edgecolor='none',
         height=0.6,
         align='center')

# Set the y ticks to arrange based on number of categories
plt.yticks(-np.arange(len(categories)),
           ['Category {}'.format(x) for x in categories])

# Add plot title
plt.title("Average Transaction Amount")
;
```

```python
## Simple Vertical Bar Chart ##
years = np.arange(2012, 2015, 1)
labels = ['2012', '2013', '2014']
values = [2, 5, 9]

plt.figure()

plt.bar(years, values)

# Set x ticks with label and rotation
plt.xticks(years, ('2012', '2013', '2014'), rotation='vertical')

# Add a y label
plt.ylabel('Values')
;
```

```python
## More Complex Vertical Bar Chart ##
years = np.arange(2012, 2015)
category1_values = [2, 5, 9]
category2_values = [7, 6, 3]

plt.figure()

plt.bar(years - 0.2,
        category1_values,
        color='gray',
        edgecolor='black',
        width=0.4,
        align='center',
        label='Year 1')

plt.bar(years + 0.2,
        category2_values,
        color='red',
        edgecolor='black',
        width=0.4,
        align='center',
        label='Year 2')

plt.xticks(years, [str(year) for year in years])

plt.legend()
;
```

Subplots are useful for comparing multiple categories against a single variable. For example, we can compare the count of stock tickers owned between different age groups from 2011 to 2017.

## EY Use Case:  Anti-Money Laundering Threshold Setting and Tuning 
---

In EY's Regulatory Technology practice, it is common for clients to ask for a distribution plot of its customers or accounts based on certain transaction patterns and then set a threshold to signal risky transactional behavior. A common metric to plot is aggregate cash payments since paying off accounts with cash is more risky. Typically, thresholds to evaluate such risky behavior are set at the 84th, 90th, and 97th percentiles - but, as we will see in this case study, sometimes these thresholds  do not capture productive activity. 

The distribution plot itself represents all customers in the sample and their aggregate cash payments for the lookback period and the "X"s represent customers that were productive, or signaled as money laundering. We plot threshold lines at corresponding percentiles and then a BAU threshold. We will use a combination of what we've learned in previous sections in this case study including pandas, scatter plot points, lines, and a distribution plot.

```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 


# Create the data frame
df = pd.read_excel("cp_short_strategic_mockdata.xlsx")
aggr_values_list = df['aggr_cash_payments'].tolist()


# Load dataframe into histogram
plt.figure()
plt.title("Productivity Distribution for aggr_cash_payments")
ax = sns.distplot(aggr_values_list)
### Get density plot coordinates
x,y = ax.lines[0].get_data()

# Add productive alerts 
int_df = pd.read_excel("cp_short_interim_mockdata.xlsx")
prod_df = int_df[int_df.prod_flag == 1]
prod_pts = prod_df['aggr_cash_payments']
prod_pts = list(prod_pts)
prod_ht = [np.interp(pt,x,y) for pt in prod_pts]
ax.scatter(prod_pts, prod_ht, c='red', s=50, marker='x', zorder=1000)
ax.set_ylabel("Density")
ax.set_ylim(ymin=0)
ax.set_xlabel("Aggr_cash_payments")

# Add BAU threshold
input_df = pd.read_excel("scenario_threshold_configuration.xlsx")
thresh_l0 = input_df['thresh_l0'].where(input_df['scnro_param']=='aggr_cash_payments').dropna().iloc[0]
plt.axvline(x=thresh_l0,color="black",label="BAU threshold")

# Plot threshold percentiles
thresh_low = input_df['thresh_perc_low'].where(input_df['scnro_param']=='aggr_cash_payments').dropna().iloc[0]
thresh_low_val = df['aggr_cash_payments'].quantile(float(thresh_low/100))
thresh_med = input_df['thresh_perc_med'].where(input_df['scnro_param']== 'aggr_cash_payments').dropna().iloc[0]
thresh_med_val = df['aggr_cash_payments'].quantile(float(thresh_med/100))
thresh_high = input_df['thresh_perc_high'].where(input_df['scnro_param']=='aggr_cash_payments').dropna().iloc[0]
thresh_high_val = df['aggr_cash_payments'].quantile(float(thresh_high/100))
low_val = int(thresh_low)
med_val = int(thresh_med)
high_val = int(thresh_high)

### Low
plt.axvline(x=thresh_low_val,color="yellow",label= str(low_val) + " percentile")
num_prod_low = sum(i < thresh_low_val for i in prod_pts)
prod_pct_low = round(num_prod_low/len(prod_pts)*100,1)
print (str(prod_pct_low) + "% of productive alerts fall below " + str(low_val) + "th percentile")

### Med
plt.axvline(x=thresh_med_val,color="orange",label= str(med_val) + " percentile")
num_prod_med = sum(i < thresh_med_val for i in prod_pts)
prod_pct_med = round(num_prod_med/len(prod_pts)*100,1)
print (str(prod_pct_med) + "% of productive alerts fall below " + str(med_val) + "th percentile")

### High 
plt.axvline(x=thresh_high_val,color="red",label= str(high_val) + " percentile")
num_prod_high = sum(i < thresh_high_val for i in prod_pts)
prod_pct_high = round(num_prod_high/len(prod_pts)*100,1)
print (str(prod_pct_high) + "% of productive alerts fall below " + str(high_val) + "th percentile")

#Legend
plt.legend(loc=3,fontsize=8,scatterpoints=1)
plt.tight_layout()
```

Now that we've plotted this visualization, it's clear the the high percentile cutoffs are insufficient to capture productive activity which is consistent with the BAU threshold at $10,000 in aggregate cash payments. This process is repeatable for all AML scenarios to visualize productive activity against a distrubtion.

## Visualizing Relationships: Regression Plots
---

Regression plots are key tools to visualize a relationship between two variables. In client work, accompanying statistical relationship data (i.e. R-squared) with a regression plot helps send a clearer message. 

The Seaborn package has visually appealling regression plot features which include, but are not limited to:
- Confidence interval bands (configuarable)
- Fitting higher order polynomial regressions and log(x)
- Placing continuous variables into discrete bins
- Residual Plots
- Plotting multiple linear regressions (in same plot and in multiple subplots)

We will first start with simple linear regression plots and how to configure parameters and fit non-linear shapes to the data using Seaborn's "regplot." Then move on to discuss how to visualize residuals for further analysis with "residplot." Finally, we explore "lmplot," a convenient interface to fit regression models across conditional subsets of a dataset.

**Seaborn commonly used "regplot" Parameters-**
_x, y:_ string, series, or vector array. 

_data:_ DataFrame

_x_ _estimator:_ callable that maps vector -> scalar, optional. Apply this function to each unique value of x and plot the resulting estimate. This is useful when x is a discrete variable. If x_ci is given, this estimate will be bootstrapped and a confidence interval will be drawn.

_x_ _bins:_ int or vector, optional. Bin the x variable into discrete bins and then estimate the central tendency and a confidence interval. This binning only influences how the scatterplot is drawn; the regression is still fit to the original data. This parameter is interpreted either as the number of evenly-sized (not necessary spaced) bins or the positions of the bin centers. When this parameter is used, it implies that the default of x_estimator is numpy.mean.

_ci:_ int in [0, 100] or None, optional. Size of the confidence interval for the regression estimate. This will be drawn using translucent bands around the regression line. The confidence interval is estimated using a bootstrap; for large datasets, it may be advisable to avoid that computation by setting this parameter to None.

_{x,y}_ _jitter:_ floats, optional. Add uniform random noise of this size to either the x or y variables. The noise is added to a copy of the data after fitting the regression, and only influences the look of the scatterplot. This can be helpful when plotting variables that take discrete values.

_label:_ string. Label to apply to ether the scatterplot or regression line (if scatter is False) for use in a legend.

_color:_ matplotlib color

_marker:_ matplotlib marker code

_ax:_ matplotlib Axes, optional. Axes object to draw the plot onto, otherwise uses the current Axes.

### Conditioning on Categorical Variables: Seaborn's "lmplot"
lmplot() combines regplot() with FacetGrid, providing an easy interface to show a linear regression on faceted plots that allow you to explore interactions with up to three additional categorical variables.

**Seaborn commonly used "lmplot" Parameters-**
_x, y:_ strings, optional

_data:_ DataFrame

_hue, col, row:_ strings. Variables that define subsets of the data, which will be drawn on separate facets in the grid. See the *_order parameters to control the order of levels of this variable.

_palette:_ palette name, list, or dict, optional. Colors to use for the different levels of the hue variable. Should be something that can be interpreted by color_palette(), or a dictionary mapping hue levels to matplotlib colors. For more color palettes and settings see: https://seaborn.pydata.org/tutorial/color_palettes.html

_markers:_ matplotlib marker code or list of marker codes, optional. Markers for the scatterplot. If a list, each marker in the list will be used for each level of the hue variable.

```python
# Comparing Scores and Attendance based on whether or not the Patriots win
# Setting confidence interval to 50% (default is 95%)
sns.lmplot(x="Winning Score", y="Attendance", hue="Pats Win", data=df2, ci=50);

# We see it doesn't make sense to categorize based on whether Pat's win or not
```
